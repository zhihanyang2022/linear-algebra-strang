```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.random.uniform(0, 10, size=(100, ))
ones = np.ones(100)
X = np.vstack([xs, ones]).T

ys = xs * 1.5 + 4.5 + np.random.normal(0, 2, size=(100, ))
```

```python
a, b = np.linalg.inv(X.T @ X) @ X.T @ ys
a, b
```

```python
xs.mean(), ys.mean()
```

```python
xs2 = ys
ys2 = xs
X2 = np.vstack([xs2, ones]).T
```

```python
a2, b2 = np.linalg.inv(X2.T @ X2) @ X2.T @ ys2
a2, b2
```

```python
fig = plt.figure(figsize=(5, 5))

plt.scatter(xs, ys, s=5)
plt.plot(xs, xs * a + b, color="blue", linewidth=1, label="y on x")
plt.xlim(0, 25)
plt.ylim(0, 25)

plt.scatter(xs2, ys2, s=5)
plt.plot(ys * a2 + b2, ys, color="red", linewidth=1, label="x on y")
plt.xlim(0, 25)
plt.ylim(0, 25)

plt.plot(np.linspace(0, 25, 10), np.linspace(0, 25, 10), color="gray", linestyle="--")

plt.legend()

plt.show()
```
