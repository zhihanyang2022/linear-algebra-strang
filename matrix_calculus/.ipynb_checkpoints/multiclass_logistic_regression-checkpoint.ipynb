```python
from functools import partial
import jax
import jax.numpy as jnp
from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt
```

## Dataset

```python
X, y = make_blobs(n_samples=1000, centers=10, cluster_std=2, n_features=2, center_box=(-20, 20), random_state=42)
X = jnp.array(X)
X -= X.mean(axis=0)
X /= X.std(axis=0)
Y = jnp.eye(10)[y]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=5)
plt.show()
```

## Autodiff

```python
def nll(w, X, Y):
    """
    X: (N, D)
    W: (D, C)
    Y: (N, C)
    """
    D = X.shape[-1]
    W = w.reshape(D, -1, order="F")
    return jnp.sum(jnp.sum(Y * (X @ W), axis=1) - jnp.log(jnp.sum(jnp.exp(X @ W), axis=1)))
```

```python
key = jax.random.PRNGKey(42)
W = jax.random.normal(key, (2, 10))
w = W.flatten("F")
```

```python
jax.grad(nll)(w, X, Y)
```

## Manual diff

```python
@partial(jax.vmap, in_axes=(0, 0, 0), out_axes=0)
def per_example_grad_wrt_w(, y, x):
    """Per example gradient of the loss function"""
    return jnp.kron( - y, x)
```

```python
M = jax.nn.softmax(X @ W, axis=1)
per_example_grad_wrt_w(M, Y, X).sum(axis=0)
```
